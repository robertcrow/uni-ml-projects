```python
import os 
import cv2

images = []

dir="C:/Users/Robert Crow/Documents/ICM/uni-ml-projects/sign_language/data/Dataset"
for subdir in os.listdir(dir):
    for filename in os.listdir(dir + '/' + subdir + '/'):
        path = dir + '/' + subdir + '/' + filename
        img = cv2.imread(path, 0)
        cv2.imshow('image', img)
        if img is not None:
            images.append(img)

        
```

```python
plt.imshow(images[0])
plt.show()
```

```python
# resizing

width = 40
height = 40
dim = (width, height)
resized_images = []

for img in images:
    resized = cv2.resize(img, dim, interpolation = cv2.INTER_AREA)
    resized_images.append(resized)
```

```python
# greyscaling images

grayscaled_images = []

for img in resized_images:
    grayscaled = np.dot(img[...,:3], [0.2989, 0.5870, 0.1140])
    grayscaled_images.append(grayscaled)
```
